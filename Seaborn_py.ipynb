```python
# Import Data Manipulation Libaries
import pandas as pd
import numpy as np

# Import Data Visualization Libaries
import matplotlib.pyplot as plt
import seaborn as sns


# Import filter Warning Libraies
import warnings
warnings.filterwarnings(action = 'ignore')

# Import logging
import logging
logging.basicConfig(level= logging.INFO,
                    filemode = 'w',
                    filename="data.log",
                    format="%(asctime)s - %(message)s - %(levelname)s",
                    force= True
                    )

# Import Style
sns.set_style("darkgrid")
```

```python
# Create DataSet X and Y

X = [10,20,30,40,5060,70,80,90,100]
Y = [23,45,56,67,78,34,56,78,85,45]


data = pd.DataFrame({
    "Age":[21,31,41,51,61,71,81,91],
    "Salary":[7000,8000.4000,6000,2000.4000,5000,8900, np.nan, np.nan],
    "Country":["India","USA","UK","India","USA","UK","UK","UK"],
    "Gender":["Male","Female","Male","Male","Female","Male","Female", np.nan]
})


# sns.lineplot
sns.lineplot(data=data,x="Age",y="Salary",
             linestyle = "--",color ='red',marker = 's',
             hue = "Gender")
plt.title("Line Plot")
plt.show()
```

```python
# scatterplot
sns.scatterplot(data=data,x="Age",y="Salary",
             marker = 's',hue = "Gender")
plt.title("Line Plot")
plt.show()
```

```python
# Barplot
sns.barplot(data=data,x="Country",y="Salary",
            hue = "Gender",orient = "h",palette= 'tab20')
plt.title("Line Plot")
plt.show()
```

```python
# kdeplot

sns.kdeplot(data=data,x="Age")
plt.axvline(data["Age"].mean(),color = "red",linestyle ='--')
plt.axvline(data["Age"].quantile(0.25),color = "orange",linestyle ='--')
plt.axvline(data["Age"].quantile(0.75),color = "orange",linestyle ='--')
plt.axvline(0.0002,linestyle = ':',color = "black")
plt.title("KDE Plot")
plt.show()
```

```python
# Heatmap : to show the correlation of two or more numerical columns with respect to target column

df = pd.DataFrame({
    "Age": [21,31,41,51,61,71,81,91],
    "Salary":[100,200,300,400,700,900,1300,1500],
    "CGPA":[8,6,8,9,4,6,7,8]
})

sns.heatmap(df.corr(),annot = True,
            vmin = -1,vmax = 1,
            center= 0,cmap ="PuBu",
            linewidths=0.02)

plt.title("Heatmap showing correlation")
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

ax.plot3D(x, y, z, 'green')
ax.set_title('3D Line Plot')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
c = x + y  # Color array based on x and y

ax.scatter(x, y, z, c=c)
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
x = np.outer(np.linspace(-2, 2, 10), np.ones(10))
y = x.copy().T
z = np.cos(x**2 + y**3)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='green')
ax.set_title('Surface Plot')
plt.show()
```

```python
def fun(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-10, 10, 40)
y = np.linspace(-10, 10, 40)
X, Y = np.meshgrid(x, y)
Z = fun(X, Y)

fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='cool', alpha=0.8)

ax.set_title('3D Contour Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

```python
from matplotlib.tri import Triangulation

def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

tri = Triangulation(X.ravel(), Y.ravel())

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(tri, Z.ravel(), cmap='cool', edgecolor='none', alpha=0.8)

ax.set_title('Surface Triangulation Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()



```

```python
R = 2
u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(-1, 1, 100)
u, v = np.meshgrid(u, v)

x = (R + v * np.cos(u / 2)) * np.cos(u)
y = (R + v * np.cos(u / 2)) * np.sin(u)
z = v * np.sin(u / 2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, alpha=0.5)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Mbius Strip')

ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])

plt.show()
```
